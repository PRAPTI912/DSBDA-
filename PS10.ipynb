```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv(r"C:\Users\Prapti\OneDrive\Desktop\DS\EDA all steps\titanic data.csv")
```

```python
print(titanic.head())
print(titanic.info())
```

```python

# ✅ Histogram
plt.figure(figsize=(6, 4))
sns.histplot(titanic['Fare'], kde=False)
plt.title("Histogram of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python

# ✅ Histplot (Recommended replacement for distplot)
plt.figure(figsize=(6, 4))
sns.histplot(titanic['Fare'], bins=30, kde=True, color='skyblue')
plt.title("Histplot of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()
```

```python
# ✅ Jointplot (Fare vs Age)
sns.jointplot(data=titanic, x='Fare', y='Age', kind='scatter')
plt.suptitle("Jointplot of Fare vs Age", y=1.02)
plt.show()
```

```python
# ✅ Rugplot
plt.figure(figsize=(6, 4))
sns.rugplot(x=titanic['Fare'], height=0.1)
plt.title("Rugplot of Ticket Fare")
plt.xlabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='Pclass', y='Fare', data=titanic)
plt.title("Barplot: Average Fare by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")
plt.show()
```
