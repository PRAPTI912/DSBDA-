```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv(r"C:\Users\Prapti\OneDrive\Desktop\DS\EDA all steps\titanic data.csv")
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic)
plt.title("Boxplot of Age by Sex and Survival")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title='Survived')
plt.show()
```

```python
# Violinplot: Age distribution by sex and survival
plt.figure(figsize=(10, 5))
sns.violinplot(x='Sex', y='Age', hue='Survived', data=titanic, split=True)
plt.title("Violinplot of Age by Sex and Survival")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title='Survived')
plt.show()
```
