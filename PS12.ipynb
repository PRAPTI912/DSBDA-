```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
```

```python
# 1. List features and their types
print("Dataset Info:")
print(iris.info())
print("\nFirst 5 rows:")
print(iris.head())
```

```python
# 2. Histogram for each feature
iris.hist(figsize=(10, 6), bins=15)
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

```python
# 3. Boxplot for each feature
plt.figure(figsize=(10, 6))
for i, column in enumerate(iris.columns[:-1], 1):  # exclude 'species'
    plt.subplot(2, 2, i)
    sns.boxplot(y=iris[column])
    plt.title(f"Boxplot of {column}")
plt.tight_layout()
plt.show()
```

```python
# 4. Demonstration of subplot using scatterplots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title("Sepal Length vs Sepal Width")

plt.subplot(1, 2, 2)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title("Petal Length vs Petal Width")
plt.tight_layout()
plt.show()
```
