```python
import pandas as pd

# Load Iris dataset
df = pd.read_csv(r"C:\Users\Prapti\OneDrive\Desktop\DSBDA labs\datasets\13Iris.csv")

# Get summary statistics for numeric variables grouped by 'species'
summary = df.groupby('species').agg(
    sepal_length_mean=('SepalLengthCm', 'mean'),
    sepal_length_median=('SepalLengthCm', 'median'),
    sepal_length_min=('SepalLengthCm', 'min'),
    sepal_length_max=('SepalLengthCm', 'max'),
    sepal_length_std=('SepalLengthCm', 'std'),
    sepal_width_mean=('SepalWidthCm', 'mean'),
    sepal_width_median=('SepalWidthCm', 'median'),
    sepal_width_min=('SepalWidthCm', 'min'),
    sepal_width_max=('SepalWidthCm', 'max'),
    sepal_width_std=('SepalWidthCm', 'std')
)

# Display the summary statistics
print(summary)
```
